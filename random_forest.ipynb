```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
pd.set_option('display.max_columns', None)
```

```python
train = pd.read_csv('./Data/train_cleaned.csv')
test = pd.read_csv('./Data/test_cleaned.csv')
target = pd.read_csv('./Data/training_set_labels.csv')
train = train.merge(target, on="id")
target = train.pop("status_group")

train['train'] = 1
test['train'] = 0

train.drop('Unnamed: 0',inplace=True,axis=1)
test.drop('Unnamed: 0',inplace=True,axis=1)

data = pd.concat([train,test])
data.head()
```

```python
data.drop(['subvillage','region_code','lga','ward','district_code','scheme_name','amount_tsh'],axis=1,inplace=True)
data.head()
```

```python
# scale numeric features

num_features=['latitude','longitude','operation_years', 'gps_height', 'population']
scaler = MinMaxScaler()

data[num_features] = scaler.fit_transform(data[num_features])
data.head()
```

```python
# factorize features for evaluations

cat_features = ['funder','installer','basin','region','public_meeting','scheme_management','permit','extraction_type','management','payment_type','water_quality','payment_type','quantity','source','waterpoint_type']

for var in cat_features:
    data[var] = pd.factorize(data[var])[0]

data.head()
```

```python
# test/train split

train_df = data[data["train"] == 1]
test_df = data[data["train"] == 0]
train_df.drop(["train"], axis=1, inplace=True)
train_df.drop(['id'],axis=1, inplace=True)
test_df.drop(["train"], axis=1, inplace=True)

id_test = test_df['id']
test_df.drop(['id'],axis=1, inplace=True)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
```

```python
forest = RandomForestClassifier(n_estimators=100, random_state = 7)
```

```python
#score = cross_val_score(classifier, train, np.ravel(target), scoring = 'accuracy',cv=5).mean()
#print("Score: {}".format(score))
train_df.info()
```

```python
forest.fit(train_df,target)
```

```python
y_pred = forest.predict(test_df)
```

```python
y_pred = pd.DataFrame(y_pred)
y_pred['id'] = id_test
y_pred.columns = ['status_group','id']
y_pred = y_pred[['id','status_group']]
```

```python
# convert into submission format

formatsub = pd.read_csv('./Data/submission_format.csv')
submission_format = pd.merge(formatsub, y_pred, on=['id'], how='inner')
submission_format.drop(['status_group_x'],axis=1,inplace=True)
submission_format.columns = ['id','status_group']
submission_format.head()
```

```python
print ("Features sorted by their score:")
print (sorted(zip(map(lambda x: round(x, 2), forest.feature_importances_), train_df), reverse=True))
```

```python
pd.DataFrame(submission_format).to_csv("./Results/submission_randomforest.csv",index=False)
```
